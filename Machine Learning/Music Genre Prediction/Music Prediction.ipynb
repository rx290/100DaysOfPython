```python
#loading Dataset
import pandas as pd
# Importing Classifier
from sklearn.tree import DecisionTreeClassifier
# Dataset splitter
from sklearn.model_selection import train_test_split
# Performance metrics for model evaluation
from sklearn.metrics import accuracy_score
# Model persistance
# from sklearn.externals import joblib
import joblib

# Decision Tree Visualization
from sklearn import tree

df = pd.read_csv('music.csv')
df
```

```python
# shape of our dataset
df.shape
```

```python
# Returns related information for data pre-processing
df.describe()
```

```python
# returns a two dimensional array of the dataset
df.values
```

```python
# Dividing dataset into two parts input and output or Training dataset and prediction dataset
x = df.drop(columns=['genre'])
x
```

```python
y = df['genre']
y
```

```python
# data split first two vars are for training and other two are for testing
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.2)
```

```python
# Creating a model

music_model = DecisionTreeClassifier()
music_model.fit(X_train,Y_train)
#Visualization
tree.export_graphviz(music_model,out_file='visual-rep-Music-Recommender.dot',
                     feature_names=['age','gender'],
                     class_names=sorted(y.unique()),
                     label='all',
                     rounded=True,
                     filled=True)

# creating a model file 
joblib.dump(music_model,'genre-prediction.joblib')
predictions = music_model.predict(X_test)
```

```python
# Evaluation
score =accuracy_score(Y_test,predictions)
score
```

```python
# loading a model file and using it to predict 
m_m = joblib.load('genre-prediction.joblib')
pred = m_m.predict([[21,1]])
pred
```
